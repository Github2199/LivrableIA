```python
#DEBUT import des bibliotheques nÃ©cessaires
import mysql.connector as msql
import pandas as pd
from mysql.connector import Error
#FIN import des bibliotheques nÃ©cessaires

#DEBUT variable
msg="ProblÃ¨me de connexion MySQL"
#FIN variable

#DEBUT Creation bdd titanicdb
try:
    conn = msql.connect(host='localhost', user='root', password='admin')
    if conn.is_connected():
        cursor = conn.cursor()
        #record = cursor.fetchone()
        print("Vous Ãªtes  sur MYSQL")
        cursor.execute('DROP DATABASE IF EXISTS titanicdb;')
        cursor.execute("CREATE DATABASE titanicdb")
        print("la base titanicdb est crÃ©ee")
    cursor.close
    conn.close     
except Error as e:
    print(msg, e)
#FIN Creation bdd titanicdb



#DEBUT creation et nettoyage de la dataframe
titanicData = pd.read_csv('c:/testPRG//titanic.csv',index_col=False)
#titanicData.head()
titanicData=titanicData.fillna(titanicData["Age"].mean())#rÃ©cupÃ©ration de la moyenne d'Ã¢ge pour remplir les valeurs vides de la colonne 'Age'
titanicData=titanicData.drop(['Cabin'], axis=1)#suppression de la colonne 'Cabin' car message d'erreur dÃ»e Ã  valeur NaN
#FIN creation et nettoyage de la dataframe

#DEBUT Creation de la table titanic et remplissage Ã  partir du dataframe, affichage des 5 premieres lignes
try:
    conn = msql.connect(host='localhost', database='titanicdb', user='root', password='admin')
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        record = cursor.fetchone()
        print("Vous Ãªtes connectÃ© Ã  la base: ", record)

        cursor.execute('DROP TABLE IF EXISTS titanic;')
        print('CrÃ©ation de la table en cours....')
        cursor.execute("CREATE TABLE titanic (PassengerId INT,Survived VARCHAR(255),Pclass VARCHAR(255),Name VARCHAR(255),Sex VARCHAR(255),Age VARCHAR(255),SibSp INT,Parch INT,Ticket VARCHAR(255),Fare VARCHAR(255),Embarked VARCHAR(255))")
        print("La table titanic est crÃ©e....")

        for i,row in titanicData.iterrows():            
            sql = "INSERT INTO titanicdb.titanic VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            conn.commit()
            # the connection is not autocommitted by default, so we must commit to save our changes 
        print("les lignes de la dataframe ont Ã©tÃ© insÃ©rÃ©es dans la table")

        cursor.execute("select * from titanicdb.titanic limit 5")
        result=cursor.fetchall()
        print("Affichage des 5 premiÃ¨res lignes de la table titanic: ")
        for i in result:
            print(result)

        cursor.close
        conn.close       

except Error as e:
    print(msg, e)
#FIN Creation de la table titanic et remplissage Ã  partir du dataframe, affichage des 5 premieres lignes
```
