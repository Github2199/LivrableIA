```python
# Importation des modules nécessaires
from sklearn.datasets import load_iris
from sklearn import tree

# Chargement du jeu de données Iris
iris = load_iris()


# Séparation des données en features (X) et étiquettes de classe (y)
X, y = iris.data, iris.target

# Initialisation d'un classificateur d'arbre de décision
clf = tree.DecisionTreeClassifier()

# Entraînement du classificateur avec les données
clf.fit(X, y)
```

```python
tree.plot_tree(clf,filled=True)
```

```python
# Importation du module graphviz
import graphviz

# Exportation de la structure de l'arbre de décision au format DOT
dot_data = tree.export_graphviz(clf, out_file=None)

# Création d'une source graphique à partir des données DOT
graph = graphviz.Source(dot_data)

# Enregistrement et rendu du graphique dans un fichier "iris.pdf"
graph.render("iris")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
n_classes = 3
plot_colors = "bry" # blue-red-yellow
plot_step = 0.02

# Choisir les attributs longueur et largeur des pétales
pair = [2, 3]

# On ne garde seulement les deux attributs
X = iris.data[:, pair]
y = iris.target

# Apprentissage de l'arbre
clf = tree.DecisionTreeClassifier().fit(X, y)

# Affichage de la surface de décision
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
plt.xlabel(iris.feature_names[pair[0]])
plt.ylabel(iris.feature_names[pair[1]])
plt.axis("tight")

# Affichage des points d'apprentissage
for i, color in zip(range(n_classes), plot_colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1], c=color, label=iris.target_names[i], cmap=plt.cm.Paired)
plt.axis("tight")
plt.suptitle("Decision surface of a decision tree using paired features")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import itertools

# Paramètres
n_classes = 3
plot_colors = "bry"  # blue-red-yellow
plot_step = 0.02

# Obtenir toutes les paires possibles d'indices d'attributs
all_pairs = list(itertools.combinations(range(len(iris.feature_names)), 2))

# Définir la taille de la figure
plt.figure(figsize=(15, 10))

# Affiche la surface de décision pour chaque paire d'attributs
for idx, pair in enumerate(all_pairs):
    # Sélectionne les attributs correspondant à la paire actuelle
    X_pair = iris.data[:, pair]

    # Apprentissage de l'arbre
    clf = tree.DecisionTreeClassifier().fit(X_pair, iris.target)

    # Affichage de la surface de décision
    x_min, x_max = X_pair[:, 0].min() - 1, X_pair[:, 0].max() + 1
    y_min, y_max = X_pair[:, 1].min() - 1, X_pair[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Affichage du subplot pour chaque paire d'attributs
    plt.subplot(2, 3, idx + 1)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
    plt.xlabel(iris.feature_names[pair[0]])
    plt.ylabel(iris.feature_names[pair[1]])

    # Affichage des points d'apprentissage
    for i, color in zip(range(n_classes), plot_colors):
        idx = np.where(iris.target == i)
        plt.scatter(X_pair[idx, 0], X_pair[idx, 1], c=color, label=iris.target_names[i], cmap=plt.cm.Paired)

    plt.axis("tight")
    plt.suptitle("Decision surface of a decision tree using paired features")
    plt.legend()

plt.tight_layout()
plt.show()
```
