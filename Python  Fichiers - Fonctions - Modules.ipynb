```python
#le PGCD de deux nombres entiers non tous deux nuls
#pgcd(44,12)=pgcd(12,8)=pgcd(8,4)=4 (44/12=3 reste 8, 12/8=4 reste 1 8/4=2 reste 0)
def pgcd(n,m):# à revoir
    div=m #3
    rest=n%m #8
    if rest!=0:
        quotient=div
        div=rest #8
        print (quotient)
        print(rest)
        pgcd(quotient,div)
    return div
    
pgcd(44,12)
#essaie d'utiliser avec round
```

```python
#histo
import pandas as pan

sx = ["M"]*5 + ["F"] * 5 + ["M"]
taille = [174,167,170,170,171,177,155,150,176,178,173]
type = []

for i in taille :

    if i < 170 : type = type+["petit"]

    else : type = type+["grand"]
mesdata = pan.DataFrame({"type":type,"sx":sx,'taille':taille})
resultat = pan.crosstab(mesdata.sx,mesdata.type, rownames = ['Sexe'], colnames = ['Type'])

print(resultat)

resultat.plot(kind="bar",rot=0)
resultat.plot()
```
